```python
def features(img):
    #initialize facial detector
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.resize(gray,(600,600))
    img = cv2.resize(img,(600,600))
    
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor('/home/abhinav/CMSC-733/Abhi1625_p2/shape_predictor.dat')
    rects = detector(gray,1)
    
    
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    for (i,rect) in enumerate(rects):
        
        shape = predictor(gray,rect)
        shape = face_utils.shape_to_np(shape)


        (x,y,w,h) = face_utils.rect_to_bb(rect)
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)


        for (x,y) in shape:
            cv2.circle(img,(x,y),2,(0,0,255),-1)
    return img
```

```python
img1 = cv2.imread('/home/abhinav/Desktop/nitinjsanket.github.io/assets/img/headshotold.jpg')
img2 = cv2.imread('/home/abhinav/CMSC-733/Abhi1625_p2/chetan.jpg')

img1 = features(img1).0.
img2 = features(img2)
```
